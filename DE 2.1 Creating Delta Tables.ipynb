```
%sql
CREATE OR REPLACE TEMP VIEW updates(id, first_name, last_name,email,gender, type) AS VALUES
  (2, "Sanju","Sampson",null,null,"update"),
  (3, "", null, null,null,"delete"),
  (7, "Blue", "Diamond",null,null, "insert"),
  (11, "Diya", "Mirza",null,null ,"update");
```

```
%sql
MERGE INTO Mock_data b
USING updates u
ON b.id=u.id
WHEN MATCHED AND u.type = "update"
  THEN UPDATE SET *
WHEN MATCHED AND u.type = "delete"
  THEN DELETE
WHEN NOT MATCHED AND u.type = "insert"
  THEN INSERT *
```
